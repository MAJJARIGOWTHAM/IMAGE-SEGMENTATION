```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from skimage import io
import matplotlib.pyplot as plt

uploader = widgets.FileUpload(
    accept='image/*',  # Accept only image files
    multiple=False  # Single file upload
)

output = widgets.Output()

def on_upload_change(change):
    if not uploader.value:
        return
    input_file = next(iter(uploader.value.values()))
    image_data = input_file['content']
    image = io.imread(image_data, plugin='imageio')

    # Process image using Mask R-CNN
    results = model.detect([image], verbose=1)
    r = results[0]

    # Clear previous output
    with output:
        clear_output(wait=True)
        # Display original image
        plt.figure(figsize=(8, 8))
        plt.imshow(image)
        plt.axis('off')
        plt.title('Original Image')
        plt.show()

        # Display segmentation masks
        visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'],
                                    class_names, r['scores'])

uploader.observe(on_upload_change, names='value')

display(uploader, output)
```

```python
uploader = widgets.FileUpload(
    accept='image/*',
    multiple=True  # Allow multiple files
)

output = widgets.Output()
gallery = widgets.HBox()  # To display a gallery of images

def process_and_display_image(image):
    results = model.detect([image], verbose=1)
    r = results[0]
    visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], class_names, r['scores'])

def on_upload_change(change):
    if not uploader.value:
        return

    images = []
    for input_file in uploader.value.values():
        image_data = input_file['content']
        image = io.imread(image_data, plugin='imageio')
        images.append(image)

    # Display all images with segmentation
    with output:
        clear_output(wait=True)
        for image in images:
            process_and_display_image(image)

uploader.observe(on_upload_change, names='value')

display(uploader, output)
```

```python
import io
from PIL import Image

def get_processed_image(image):
    results = model.detect([image], verbose=1)
    r = results[0]

    # Create a figure to hold the plot
    fig, ax = plt.subplots(1, figsize=(12, 12))
    visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], class_names, r['scores'], ax=ax)
    ax.axis('off')

    # Save the figure to a bytes buffer
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    plt.close(fig)
    buf.seek(0)
    return buf
```

```python
import base64

def get_image_data(image_buffer):
    image_data = base64.b64encode(image_buffer.read()).decode('utf-8')
    image_buffer.close()
    return image_data
```

```python
def on_upload_change(change):
    if not uploader.value:
        return

    links = []
    for input_file in uploader.value.values():
        image_data = input_file['content']
        image = io.imread(image_data, plugin='imageio')
        image_buffer = get_processed_image(image)
        image_data_base64 = get_image_data(image_buffer)

        link = f"<a href='data:image/png;base64,{image_data_base64}' target='_blank'>Open Segmented Image in New Tab</a>"
        links.append(link)

    with output:
        clear_output(wait=True)
        display(HTML('<br>'.join(links)))

uploader.observe(on_upload_change, names='value')

display(uploader, output)
```
